```python
import PyPDF2
import nltk
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
print("Upload your Resume PDF")
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

reader = PyPDF2.PdfReader(file_name)
resume_text = ""

for page in reader.pages:
    text = page.extract_text()
    if text:
        resume_text += text + " "

print("\nResume text extracted successfully!")
```

```python
jd_text = input("Paste the Job Description here:\n")
```

```python
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\d+', '', text)
    text = text.translate(str.maketrans("", "", string.punctuation))
    tokens = nltk.word_tokenize(text)
    tokens = [stemmer.stem(w) for w in tokens if w not in stop_words]
    return " ".join(tokens)
```

```python
clean_resume = preprocess_text(resume_text)
clean_jd = preprocess_text(jd_text)
```

```python
skills = [
    "python", "java", "sql", "excel", "power bi", "tableau",
    "machine learning", "deep learning", "nlp", "statistics",
    "data analysis", "data visualization", "pandas", "numpy",
    "communication", "teamwork", "problem solving"
]

def extract_skills(text):
    found = []
    for skill in skills:
        if skill.lower() in text.lower():
            found.append(skill)
    return found

resume_skills = extract_skills(resume_text)
jd_skills = extract_skills(jd_text)

missing_skills = list(set(jd_skills) - set(resume_skills))
```

```python
corpus = [clean_resume, clean_jd]

vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform(corpus)

similarity_score = cosine_similarity(vectors[0], vectors[1])[0][0]
match_percentage = round(similarity_score * 100, 2)
```

```python
print("\n===============================")
print("        MATCH RESULTS")
print("===============================\n")

print(f"üîç Resume‚ÄìJob Match Score: {match_percentage}%")

print("\n‚úÖ Skills Found in Resume:")
print(resume_skills)

print("\nüìå Skills Required in JD:")
print(jd_skills)

print("\n‚ùó Missing Skills (Improve these):")
print(missing_skills if missing_skills else "None! Good job üíõ")

print("\nüéØ Overall Recommendation:")
if match_percentage >= 80:
    print("Excellent! Your resume strongly matches this job.")
elif match_percentage >= 60:
    print("Good. Add the missing skills to increase your chances.")
else:
    print("Needs improvement. Try adding more relevant skills & projects.")
```
